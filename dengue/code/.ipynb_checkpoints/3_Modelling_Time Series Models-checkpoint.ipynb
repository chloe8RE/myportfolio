# 3.1 Time Series Modelling

For time-series models, the general process are as follows:
* Ensure time series data stationarity
* Plot correlation and autocorrelation charts
* Construct ARIMA and SARIMAX Model based on data
* Make predictions using the model

## Premodelling

Since we will be predicting dengue cases through time series modelling, let's call out the dengue_incidence df with only dengue_cases as our y-variable.

```python
dengue_incidence = dengue_wk[['dengue_cases']]
dengue_incidence.head()
```

To use time series models such as ARIMA, SARIMA or SARIMAX, the data has to be free of trend and seasonality. From the below seasonal_decomposition chart that we plotted earlier in notebook 2, we can already tell that the dengue_cases data has seaonality.

```python
decomposition = seasonal_decompose(dengue_wk['dengue_cases'], model='additive')

# Plot the decomposition
plt.figure(figsize=(10, 8))
decomposition.plot();
plt.show();
```

### Check Stationarity

The time-series data must be stationary to develop a model that is effective in forecasting future values.  Firstly, let's plot Augmented Dickey–Fuller test (ADF test), the partial correlation function (PACF), and the autocorrelation function (ACF) to determine stationarity of the data.

```python
# Check stationarity:
from pmdarima.arima import ADFTest

result = adfuller(dengue_incidence, autolag='AIC')
print(f'ADF Statistic: {result[0]}')
print(f'n_lags: {result[1]}')
print(f'p-value: {result[1]}')
for key, value in result[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')    
```

```python
# Check stationarity:
from statsmodels.tsa.stattools import adfuller

ad_fuller_result = adfuller(dengue_wk['dengue_cases'])
print(f"p-value for original:{str(ad_fuller_result[1])}")
```

##### Findings
Since p-value is 0.0036 (< 0.05), we can reject the null hypothesis (that data is not stationary) because it is statistically insignificant. Hence, we can treat the data as stationary.

### Check Seasonality

let's explore several lag intervals, so we can see from the autocorrelation whether there is any particular duration of lag that has strong correlation to the current week. This would imply the seasonal influences of our data.

Based on some research and domain knowledge, the following lag intervals were decided:
* 1 week - [because symptoms may develop after an incubation period of approximately 7 days](https://www.parkwayeast.com.sg/conditions-diseases/dengue-fever/symptoms-causes)
* 2 weeks - <mark>RETURN TO THIS</mark>
* 4 weeks
* 8 weeks
* 12 weeks
* 52 weeks - to test for any annual seasonal influence 

```python
print(f'Autocorrelation for 1 week lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=1)}')
print(f'Autocorrelation for 2 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=2)}')
print(f'Autocorrelation for 4 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=4)}')
print(f'Autocorrelation for 8 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=8)}')
print(f'Autocorrelation for 12 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=12)}')
print(f'Autocorrelation for 52 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=52)}')
```

##### Findings
- The single week autocorrelation shows that there is a strong correlation between dengue cases each week from the prior week. The 2 week autocorrelation also tells us that the 2 week lag of dengue cases also has a strong correlation to current week cases.
- Looking at week 4 onwards, it seems that the further away from the current week, the less influence that week's sales have on the current week's sales. When we reach the 52-week lag of sales, there is negative correlation.
- The findings tell us that there might be seasonal influence on dengue cases, separated by a 1 or 2 wk lag.

### ACF, PACF
Let's plot 52-wk ACF, PACF

```python
# Plot Autocorr
plot_acf(dengue_incidence, lags=52);

# Plot Partial Autocorr
plot_pacf(dengue_incidence, lags=52);
```

In the ACF plot, we observe geometric decay, indicating that the autocorrelation between the time series and its past observations decreases gradually as the lag increases. The tail off pattern in the ACF plot, suggests the need to use AutoRegression(AR). 

<mark>The order of the autoregression ("p"), it is determine by the PACF plot where there is a huge drop (in the sense that it falls into the blue range which is just the 95% significance threshold). So based on the plot, AR would be set at 1 since the drop occurs at lag =1 in the PACF **REPHRASE**</mark>

## Model 2 - SARIMAX
The earlier baseline SARIMA model did not perform well, as it was unable to account for other features, such as weather and google trends. Let's add these additional features to form our Seasonal Autoregressive Integrated Moving Average with eXogenous Regressors model (SARIMAX) model.

```python
#recap our earlier dataframe with all features
dengue_wk.head()
```

```python
#drop df and dhf, as we will be using dengue_cases as our y-variable to predict
dengue_wk.drop(columns=['df', 'dhf'], inplace=True)
```

- Our target variable to forecast: `dengue_cases`
- Let's incorporate our weather and google_trends (exogenous features) `X`. We'll lag them and include into our modeling to see if it improves our model's performance
    - For weather features, based on insights from our EDA, let's feature engineer new columns that lag the data by `1 week`, and `1 month`

##### Lagging the weather features

```python
dengue_wk.columns
```

```python
# weather columns to lag
weather_col = ['highest_30min_rainf', 'highest_60min_rainf', 'highest_120min_rainf', 'daily_rainf_total', 'mean_temp', 'max_temp', 'min_temp', 'mean_wind_speed', 'max_wind_speed']

for col in weather_col:
    dengue_wk[col+"_prev_wk"] = dengue_wk[col].shift(1) 
    dengue_wk[col+"_prev_mth"] = dengue_wk[col].shift(4) 
    dengue_wk = dengue_wk.dropna() # drop NaN rows occuring from shifting operations above

dengue_wk.head() # compare the actual and lagged columns
```

Let's drop 'highest_30min_rainf', 'highest_60min_rainf', 'highest_120min_rainf' as these 3 features are multicollinear to the 'daily_rainf_total' (total daily rainfall) feature:

```python
#drop 'highest_30min_rainf', 'highest_60min_rainf', 'highest_120min_rainf'
dengue_wk.drop(columns=['highest_30min_rainf', 'highest_60min_rainf', 'highest_120min_rainf'], inplace=True)
```

```python
# weather columns to lag
weather_col = ['daily_rainf_total', 'mean_temp', 'max_temp', 'min_temp', 'mean_wind_speed', 'max_wind_speed']

for col in weather_col:
    dengue_wk[col+"_prev_wk"] = dengue_wk[col].shift(1) 
    dengue_wk[col+"_prev_mth"] = dengue_wk[col].shift(4) 
    dengue_wk = dengue_wk.dropna() # drop NaN rows occuring from shifting operations above

dengue_wk.head() # compare the actual and lagged columns
```

```python
# Adding Exogenous variables to our model
X = dengue_wk.drop(columns=['dengue_cases'])
X_train = X[:'2021']
X_test = X['2022':]

# same steps as earlier to split `y`
y = dengue_wk['dengue_cases']
y_train = y[:'2021']
y_test = y['2022':]

#to ensure the split is correct (we know 2021-2022 consists of 104 rows)
print(X_test.shape) 
```

```python
# Fit Auto ARIMA
# all parameter declarations are SAME as SARIMA
sarimax_model = pm.AutoARIMA(start_p=1, max_p=1, 
                           d=None,    # find optimum value of d automatically
                           start_q=0, max_q=8,
                           start_P=0, max_P=8,
                           D=None,
                           start_Q=0, max_Q=8,
                           m=12, 
                           seasonal=True,
                           trace=True, # Print values for each fit in the grid search
                           random_state=20,
                           n_fits=50
                          )

sarimax_model.fit(y_train, X_train) # we were only fitting purely on y_train before, but now we are fitting on y_train and X_train to include exogenous features
```

```python
plot_forecast(y_train, y_test, sarimax_model, X_test=X_test)
```
