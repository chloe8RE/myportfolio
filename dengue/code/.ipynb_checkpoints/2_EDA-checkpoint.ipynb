### Seasonal Decomposition for Dengue Cases

```python
decomposition = seasonal_decompose(dengue_wk['dengue_cases'], model='additive')

# Plot the decomposition
plt.figure(figsize=(10, 8))
decomposition.plot();
plt.show();
```

From the seasonal decomposition plot, we are able to tell that there is seasonality to the number of dengue cases, and each season is in line with the duration of a year. Later on, when working with time series models, we will need to make sure to remove seasonality to achieve stationarity to forecast effectively.

## Weather Trends
* Plot seasonal decomposition charts
* Plot Temperature vs Wind
* Plot Temperature vs Rainfall

Let's try plotting seasonal decomposition charts for rainfall and temperature. This is a naive decomposition, but would be able to give us a preliminary gauge of the trends and seasonality for the weather factors. 

The results are obtained by first estimating the trend by applying a convolution filter to the data. The trend is then removed from the series and the average of this de-trended series for each period is the returned seasonal component. I decided to use the additive model because the seasonal variation is relatively constant over time.


```python
#create function for seasonal decomposition plots
def seasonal_decomposition(df, column):
    # Perform seasonal decomposition 
    decomposition = seasonal_decompose(df[column], model='additive')

    # Plot the decomposition
    plt.figure(figsize=(10, 8))
    decomposition.plot()
    plt.show();
```

```python
seasonal_decomposition(dengue_wk, 'daily_rainf_total');
```

```python
seasonal_decomposition(dengue_wk, 'mean_temp');
```

## Plot Dual Line Graphs

Create function to plot dual line graphs, to compare various weather conditions with no of dengue cases.

```python
#create function to plot dual line graphs
def dual_line_plot(ax1_col, label1, suptitle):
    color_1 = "steelblue"
    color_2 = "salmon"

    fig, ax1 = plt.subplots(figsize=(8, 5))
    ax2 = ax1.twinx()
    
    ax1.plot(dengue_wk.index, dengue_wk[ax1_col], color=color_1, lw=1, alpha=0.7)  
    ax2.plot(dengue_wk.index, dengue_wk['dengue_cases'], color=color_2, lw=1.2)  
    
    ax1.set_xlabel("Date")
    ax1.set_ylabel(label1, color=color_1, fontsize=10)
    ax1.tick_params(axis="y", labelcolor=color_1)
    
    ax2.set_ylabel("Dengue Case Count", color=color_2, fontsize=10)
    ax2.tick_params(axis="y", labelcolor=color_2)
    
    # Remove background grid
    ax1.grid(axis='x')
    ax2.grid(False)
    
    fig.suptitle(suptitle, fontsize=12)
    
    plt.show()
```

### Temperature vs No of Dengue Cases
Plot Temperature vs dengue cases. In this case, we use average mean temperature.

```python
dual_line_plot('mean_temp', "Temperature (°C)", "Temperature VS Dengue Cases")
```

### Average Daily Total Rainfall vs No of Dengue Cases

```python
dual_line_plot('daily_rainf_total', "Average Daily Total Rainfall (mm)", "Average Daily Total Rainfall VS Dengue Cases")
```
