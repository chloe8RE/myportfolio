# Time Series Modelling

For time-series models, the general process is as follows:
* Ensure time series data stationarity
* Plot correlation and autocorrelation charts
* Construct ARIMA and SARIMAX Model based on data
* Make predictions using the model

## Premodelling

Since we will be predicting dengue cases through time series modelling, let's call out the dengue_incidence df with only dengue_cases as our y-variable.

```python
dengue_incidence = dengue_wk[['dengue_cases']]
dengue_incidence.head()
```

To use time series models such as ARIMA, SARIMA or SARIMAX, the data has to be free of trend and seasonality. From the below seasonal_decomposition chart that we plotted earlier in notebook 2, we can already tell that the dengue_cases data has seasonality.

```python
decomposition = seasonal_decompose(dengue_wk['dengue_cases'], model='additive')

# Plot the decomposition
plt.figure(figsize=(10, 8))
decomposition.plot();
plt.show();
```

The trend component in the seasonal decomposition plot lacks of a well-defined pattern, as it alternates between upward and downward movements. The seasonal trend showcases a recurring pattern that peaks around the middle of each year, indicating seasonality and consistent fluctuations.

### Check Stationarity

The time-series data must be stationary to develop a model that is effective in forecasting future values.  Firstly, let's plot Augmented DickeyFuller test (ADF test), the partial correlation function (PACF), and the autocorrelation function (ACF) to determine stationarity of the data.

```python
# Check stationarity:
from statsmodels.tsa.stattools import adfuller

ad_fuller_result = adfuller(dengue_wk['dengue_cases'])
print(f"p-value for original:{str(ad_fuller_result[1])}")
```

Let's plot 52-wk ACF, PACF:

```python
# Plot Autocorr
plot_acf(dengue_incidence, lags=52);

# Plot Partial Autocorr
plot_pacf(dengue_incidence, lags=52);
```

##### Findings
From the ADF test, p-value is 0.0036 (< 0.05). This tells us we can reject the null hypothesis (that data is not stationary) because it is statistically insignificant. However, the ACF plot tells us otherwise.

In the ACF plot, we observe geometric decay. This tells us there is a need for differencing since the autocorrelation values remain significant and slowly decay as the lag increases. The data still has some seasonality or trend present, and we will carry out differencing after doing our train test split. The tailing off pattern also suggests the suitability of AutoRegression (AR) modeling. 

### Check Seasonality

let's explore several lag intervals, so we can see from the autocorrelation whether there is any particular duration of lag that has strong correlation to the current week. This would imply the seasonal influences of our data.

Based on some research and domain knowledge, the following lag intervals were decided:
* `1 week / 2weeks` - accounts for  the incubation period of the dengue virus within a human host. Dengue symptoms typically start to manifest around [3 to 14 days](https://www.parkwayeast.com.sg/conditions-diseases/dengue-fever/symptoms-causes) after a person is bitten by an infected mosquito.
* `4 weeks` - captures the impact of mosquito breeding and maturation on dengue transmission
* `8 weeks`
* `12 weeks`
* `52 weeks` - to test for any annual seasonal influence 

```python
print(f'Autocorrelation for 1 week lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=1)}')
print(f'Autocorrelation for 2 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=2)}')
print(f'Autocorrelation for 4 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=4)}')
print(f'Autocorrelation for 8 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=8)}')
print(f'Autocorrelation for 12 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=12)}')
print(f'Autocorrelation for 52 weeks lag interval: {dengue_incidence["dengue_cases"].autocorr(lag=52)}')
```

##### Findings
- The single week autocorrelation shows that there is a strong correlation between dengue cases each week from the prior week. The 2 week autocorrelation also tells us that the 2 week lag of dengue cases also has a strong correlation to current week cases.
- Looking at week 4 onwards, it seems that the further away from the current week, the less influence that week's sales have on the current week's sales. When we reach the 52-week lag of sales, there is negative correlation.
- The findings tell us that there might be seasonal influence on dengue cases, separated by a 1 or 2 wk lag.

## Baseline Model 1 - SARIMA

From the preliminary 52 wk ACF plot, we observed significant autocorrelation values at multiple lags, suggesting that the data still has some seasonality or trend present. Below, we tested differencing the data. Only upon the second differencing, did we achieve stationarity (from the ACF plot, we can see most of the lags now fall within the 95% significance threshold).

```python
# diff() gets delta between every month vs previous month - achieves stationarity
y_train.diff().diff().head()
```

```python
# drop null from 1st month with dropna()
y_train.diff().diff().dropna().head()
```

```python
# Generate ACF, PACF plot on training data.
plot_acf(y_train.diff().diff().dropna());
plot_pacf(y_train.diff().diff().dropna());
```


To determine the order of the autoregression ("p"), we use the Partial Autocorrelation Function (PACF) plot to gain insights. The PACF plot shows a significant drop at lag 2, suggesting an AR order of 2 for modeling, as it falls within the 95% significance threshold.

```python
# Fit Auto SARIMA
sarima_model = pm.AutoARIMA(
                           # same parameters as Auto ARIMA
                           start_p=1, max_p=8, 
                           d=None,   
                           start_q=0, max_q=8, 
                           
                           start_P=1, max_P=1, 
                           D=None,    
                           start_Q=0, max_Q=8, 
                           m=52, 
                           seasonal=True, 
                           
                           # same parameters as Auto ARIMA
                           trace=True, 
                           random_state=20,
                           n_fits=50
                          )

sarima_model.fit(y_train)
```

##### Inferences from Auto SARIMA model
- `AR` best hyperparameter:`p` - 1
- `I` best hyperparameter:`d` - 2 (2nd order difference was required to make this time-series stationary)
- `MA` best hyperparameter: `q` - 2

```python
# trained model summary stats
sarima_model.summary()
```

```python
compiled_functions.plot_forecast(y_train, y_test, sarima_model, "SARIMA Prediction of Dengue Cases - 1 year")
```

## Model 2 - SARIMAX
The earlier baseline SARIMA model did not perform well, as SARIMA only relies on past observations of the target variable itself, and it is unable to account for other exogenous features, such as weather and google trends. These features can variables can provide additional information and context, leading to better forecasting accuracy. Let's add these additional features to form our Seasonal Autoregressive Integrated Moving Average with eXogenous Regressors model (SARIMAX) model.

```python
#recap our earlier dataframe with all features
dengue_wk.head()
```

```python
#drop df and dhf, as we will be using dengue_cases as our y-variable to predict
dengue_wk.drop(columns=['df', 'dhf'], inplace=True)
```

- Our target variable to forecast: `dengue_cases`
- Let's incorporate our weather and google_trends (exogenous features) `X`. We'll lag them and include into our modeling to see if it improves our model's performance
    - For weather features, based on insights from our EDA, let's feature engineer new columns that lag the data by `1 week`, and `1 month`

##### Lagging the weather features

```python
dengue_wk.columns
```

Let's drop 'highest_30min_rainf', 'highest_60min_rainf', 'highest_120min_rainf' as these 3 features are multicollinear to the 'daily_rainf_total' (total daily rainfall) feature:

```python
#drop 'highest_30min_rainf', 'highest_60min_rainf', 'highest_120min_rainf'
dengue_wk.drop(columns=['highest_30min_rainf', 'highest_60min_rainf', 'highest_120min_rainf'], inplace=True)
```

```python
# weather columns to lag
weather_col = ['daily_rainf_total', 'mean_temp', 'max_temp', 'min_temp', 'mean_wind_speed', 'max_wind_speed']

for col in weather_col:
    dengue_wk[col+"_prev_wk"] = dengue_wk[col].shift(1) 
    dengue_wk[col+"_prev_mth"] = dengue_wk[col].shift(4)
    dengue_wk = dengue_wk.dropna() # drop NaN rows occuring from shifting operations above

dengue_wk.head() # compare the actual and lagged columns
```

```python
# Adding Exogenous variables to our model
X = dengue_wk.drop(columns=['dengue_cases'])
X_train = X[:'2021']
X_test = X['2022':]

# same steps as earlier to split `y`
y = dengue_wk['dengue_cases']
y_train = y[:'2021']
y_test = y['2022':]

#to ensure the split is correct (we know 2021-2022 consists of 104 rows)
print(X_test.shape) 
```

```python
sarimax_model = pm.AutoARIMA(start_p=1, max_p=8, 
                        d=None,    # find optimum value of d automatically
                        start_q=0, max_q=8,
                        start_P=0, max_P=8,
                        D=None,
                        start_Q=0, max_Q=8,
                        m=52, 
                        seasonal=True,
                        trace=True, # Print values for each fit in the grid search
                        random_state=20,
                        n_fits=50
                        )

sarimax_model.fit(y_train, X_train) 
```

```python
compiled_functions.plot_forecast(y_train, y_test, sarimax_model, "SARIMAX Prediction of Dengue Cases - 1 year", X_test=X_test)
```

## Model 3 - SARIMAX
While our earlier SARIMAX model was able to predict the 2022 dengue spike relatively accurately, we will now refine our production model to predict just 3 months ahead. <br>

let's explore running 3 month predictions for the year of 2022:
* Jan - Mar
* Apr - Jun
* Jul - Aug
* Sep - Dec

### Creating functions

```python
def sarimax_ttsplit(train_end, test_start, test_end, X, y):
    X_train = X[:train_end]
    X_test = X[test_start:test_end]
    
    y_train = y[:train_end]
    y_test = y[test_start:test_end]
    
    # To ensure the split is correct
    print(X_test.shape)
    
    return X_train, X_test, y_train, y_test
```

```python
def fit_sarimax(y_train, X_train):
    sarimax_model = pm.AutoARIMA(
        start_p=1, max_p=8, 
        d=None, start_q=0, max_q=8,
        start_P=0, max_P=8,
        D=None, start_Q=0, max_Q=8,
        m=52, seasonal=True,
        trace=True, random_state=20,
        n_fits=50
    )

    sarimax_model.fit(y_train, exogenous=X_train) 

    return sarimax_model
```

### Jan - Mar 2022

```python
#train test split for mentioned duration
X_train_jan_mar, X_test_jan_mar, y_train_jan_mar, y_test_jan_mar = sarimax_ttsplit('2021-12', '2022-01', '2022-03', X, y)
```

```python
sarimax_model_jan_mar = fit_sarimax(y_train_jan_mar, X_train_jan_mar)
```

```python
compiled_functions.plot_forecast(y_train_jan_mar, y_test_jan_mar, sarimax_model_jan_mar, "SARIMAX Prediction of Dengue Cases - Jan to Mar 2022", X_test=X_test_jan_mar)
```
