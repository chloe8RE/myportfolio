```python
# Set YouTube Data API key to scrape Youtube video comments.
API_KEY = 'AIzaSyBVl-M4kffoc2ifiWSnwT-qkgLFWFtOa80'

# Create YouTube Data API client
youtube = build('youtube', 'v3', developerKey=API_KEY)

def get_video_comments(video_id):
    comments_data = []

    # Get the first page of comments
    response = youtube.commentThreads().list(
        part='id,snippet,replies',
        videoId=video_id,
        textFormat='plainText',
        maxResults=100        # Maximum number of comments per request
    ).execute()

    while response:
        for item in response['items']:
            comment = item['snippet']['topLevelComment']['snippet']['textDisplay']
            comment_id = item['snippet']['topLevelComment']['id']
            author = item['snippet']['topLevelComment']['snippet']['authorDisplayName']
            likes_count = item['snippet']['topLevelComment']['snippet']['likeCount']
            comments_data.append({'comment': comment, 'comment_id': comment_id, 'author': author, 'likes': likes_count})

            # Check if the comment has replies
            if 'replies' in item:
                for reply_item in item['replies']['comments']:
                    reply_comment = reply_item['snippet']['textDisplay']
                    reply_commentid = reply_item['id']
                    reply_author = reply_item['snippet']['authorDisplayName']
                    reply_likes = reply_item['snippet']['likeCount']
                    comments_data.append({'comment': reply_comment, 'comment_id': reply_commentid, 'author': reply_author, 'likes': reply_likes})


        # Check if there are more comments
        if 'nextPageToken' in response:
            response = youtube.commentThreads().list(
                part='id,snippet,replies',
                videoId=video_id,
                textFormat='plainText',
                maxResults=100,
                pageToken=response['nextPageToken']
            ).execute()
        else:
            break

    return comments_data

# The video ID you want to fetch comments from
video_id = 'bk-nQ7HF6k4'
comments_data = get_video_comments(video_id)

# print dataframe for 23k rows and 4 columns
df = pd.DataFrame(comments_data)
```

### 4.5 Cleaning

```python
# Clean the translated comments to prepare for modelling
def clean_text(x):
    x = x.encode('ascii','ignore').decode()  # remove emojis
    x = re.sub('[%s]'%re.escape(string.punctuation),' ',x)  # remove punctuations
    x = re.sub(r'https*\S+',' ',x)  # remove URL
    x = re.sub(r'\'\w+',' ',x)  # remove ticks
    x = re.sub(r'#\S+',' ',x)  # remove hashtags
    x = re.sub(r'@\S+',' ',x)  # remove mentions
    x = re.sub(r'\w*\d+\w*',' ',x)   # remove numbers
    x = re.sub(r'\s{2,}',' ',x)   # remove extra whitespaces that occurs more than 2 times
    return x

df['cleaned_comment'] = df['trans_comment'].apply(clean_text)
df
```

```python
# Noticed there is inconsistent format (uppercase), hence change to lowercase for EDA and processing.
df['cleaned_comment'] = df['cleaned_comment'].apply(str.lower)
df
```

> Noises have been cleared from the dataframe (corpus). Punctuations, URLs, ticks, hastags, mentions and numbers are now removed.  Lowercasing has also been applied to the comment column. I can now prepare the text data for empty cells checking, which is common after removal of emojis and special character.  These will be replaced with np.NaN then dropped from dataframe.

## 5. Preliminary Sentiment Classifier

> Vader (Valence Aware Dictionary and Sentiment Reasoner) is chosen because this classifier tool is fundamentally designed for easy classifying sentiments gathered from social media, catered for buzzwords that online netizens would use to express their feelings or opinions that might be vastly different from other platforms (e.g. product reviews on Amazon/Lazada, or movies reviews on IMDb).  Vader takes into account emojis, exclamation marks and capital letter to classify sentiments hence it is useful for my case study of Youtube videos comments.

> Vader is used to calculate the negative, positive, neutral and compound scores. I added label of positive or negative by setting a threshold of 0. Compound score > 0 is positive and 0 and below is classified as negative.

> I take a sample of 100 from the raw datasets to input my own human classifier, and validate against Vader scores.  This is to ensure Vader assessment methodolody from machine learning aligned with human's before fitting into modelling to devise reliable classification and predictions.

```python
df = pd.read_csv('data/cleaned_comments.csv')
```

```python
# Instantiate Sentiment Intensity Analyzer
sia = SentimentIntensityAnalyzer()
```

```python
# Define a function to apply Vader sentiment scoring to every row of comments
def vader_sentiment(comment):
    sia = SentimentIntensityAnalyzer()     
    sentiment_scores = sia.polarity_scores(comment)
    return sentiment_scores


df['vader_scores'] = df['cleaned_comment'].apply(vader_sentiment)
df['vader_compound'] = df['vader_scores'].apply(lambda x: x['compound'])
df['vader_label'] = df['vader_compound'].apply(lambda score: 'positive' if score > 0 else 'negative')
df['label'] = df['vader_label'].map({'negative': 0, 'positive': 1})
```

### 6.3 Term Frequency Analysis

```python
# Start with one review:
text = df['cleaned_comment'].str.cat()

# Add custom stopwords to the STOPWORDS set, as these words dont seem relevant and will appear in both positive and negative comments
custom_stopwords = {'s','t','one','will','need','want','one','don','now','live','u','us','thing','things','even','already','really'}
modified_stopwords = STOPWORDS.union(custom_stopwords)

# Remove common stop words using the WordCloud's STOPWORDS set
stopwords = set(modified_stopwords)

# Create and generate a word cloud image:
wordcloud = WordCloud(stopwords=stopwords, width=2000, height=1200, background_color='white').generate(text)

# Display the generated image:
fig = plt.figure(figsize=(18,4), facecolor='white')
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title('Top 100 Most Common Words', fontsize=20)
plt.tight_layout(pad=0)
plt.show()
```

### 6.4 Top 20 Words

```python
# Extract the 'text' data from Positive Comments and concatenate into a string
text_data = df['cleaned_comment'].str.cat()

# Pre-processing: lowercasing and remove punctuation
text_data = text.lower()
text_data = re.sub(r'[^\w\s]', '', text_data)

# Add custom stopwords to the STOPWORDS set, as these words dont seem relevant and will appear in both positive and negative comments
custom_stopwords = {'s','t','will','u','one','even','us','need','humans'}
modified_stopwords = STOPWORDS.union(custom_stopwords)

# Remove common stop words using the WordCloud's STOPWORDS set
stopwords = set(modified_stopwords)
text_data = " ".join([word for word in text_data.split() if word not in stopwords])

# Generate the word cloud
wordcloud = WordCloud(stopwords=modified_stopwords, width=800, height=500, background_color='white').generate(text_data)

# Extract the word frequencies from the word cloud
word_frequencies = wordcloud.words_

# Convert the word frequencies to a DataFrame
word_freq_df = pd.DataFrame(list(word_frequencies.items()), columns=['Word', 'Frequency'])
word_freq_df = word_freq_df.sort_values(by='Frequency', ascending=False)

# Select the top 20 words
top_20_words = word_freq_df.head(20)
top_20_words = top_20_words.sort_values(by='Frequency', ascending=True)

# Plot the horizontal bar chart
plt.barh(top_20_words['Word'], top_20_words['Frequency'])
plt.xlabel('Frequency')
plt.ylabel('Words')
plt.title('Top 20 Words Frequency Ranking')
plt.tight_layout()
plt.show()
```
