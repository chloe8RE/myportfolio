```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Plot ACF for different differencing orders
plot_acf(y_diff2.index, ax=ax1) 
ax1.set_title("ACF - Autocorrelation", fontsize=12)

plot_pacf(y_diff2.index, ax=ax2)
ax2.set_title("PACF - Partial Autocorrelation", fontsize=12)

plt.tight_layout()

# Show the plots
plt.show()
```

```python
#Take abt 65% train, 35% test
y_train_IA=y_diff2[:174]
y_test_IA=y_diff2[174:]
```

```python
decomposition = seasonal_decompose(x=y_train_IA, model='addicitve')

# Plot the decomposition
fig = decomposition.plot()
fig.set_size_inches((12,6))
fig.tight_layout();
```

```python
# Fit Auto ARIMA
arima_model = pm.AutoARIMA(
                           start_p=0, max_p=10, 
                           d=None,   
                           start_q=0, max_q=10,    
                           # same parameters as Auto ARIMA
                           trace=True, 
                           random_state=20,
                           n_fits=50
                          )

arima_model.fit(y_train_IA)
```

```python
def evaluate_model(model, y_train, y_test):
    # Make predictions on train and test sets
    y_pred_train = model.predict(n_periods = len(y_train))
    y_pred_in_sample = model.predict_in_sample()
    y_pred_test = model.predict(n_periods = len(y_test))
    
    # Calculate RMSE and MAPE for train and test sets
    rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
    rmse_train2= np.sqrt(mean_squared_error(y_train, y_pred_in_sample))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
    mape_train = mean_absolute_percentage_error(y_train, y_pred_train)
    mape_test = mean_absolute_percentage_error(y_test, y_pred_test)
    
    # Print the results
    print('Train RMSE:', rmse_train)
    print('Train RMSE (in-sample):', rmse_train2)
    print('Test RMSE:', rmse_test)
    print('Train MAPE:', mape_train)
    print('Test MAPE:', mape_test)
```

```python
evaluate_model(arima_model, y_train_IA, y_test_IA)
```

```python
# Create helper function to help make predictions and plot the results 
# function captures the same steps as previous lesson with just the addition of `X_test` to use for SARIMAX model next

def plot_forecast(y_train, y_test, arima_model, X_test=None):
    # Generate len(y_test) number of predictions
    y_pred = arima_model.predict(n_periods = len(y_test), X=X_test) # we want as many future data point predictions as there are in y_test

    # Plot data.
    plt.figure(figsize=(30,15))

    # Plot training data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    plt.plot(y_train, color='blue', label='y_train')

    # Plot testing data. don't need to explicitly pass x-axis for chart, taken as the date index by matplotlib
    plt.plot(y_test, color='orange', label='y_test', alpha = 0.7)

    # Plot predicted test values. as shared previously, there are no index in y_pred 
    # so, we need to explicitly pass the `x-axis` (dates) to compare for same timeframe as y_test and access y_pred vs actual y_test
    # we can use y_test.index to get the index of y_test as confirmed earlier in y_test.head()
    plt.plot(y_pred, color='green', label='y_pred', alpha = 0.9, ls = '--')

    y_pred_in_sample = arima_model.predict_in_sample()
    y_pred_test = arima_model.predict(n_periods = len(y_test))
    
    # Find the MAPE (mean abs percentage error) of the predictions
    # https://scikit-learn.org/stable/modules/generated/sklearn.metrics.mean_absolute_percentage_error.html
    rmse_train= np.sqrt(mean_squared_error(y_train, y_pred_in_sample))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

    plt.title(label = f'Sales Predictions\n test RMSE: {rmse_test:.2f} \n train RMSE: {rmse_train:.2f}', fontsize=16) #\n is just for new line to print MAPE 1 line after header 'Sales Predictions', while .2f rounds MAPE to 2 decimals
    plt.legend(fontsize = 20, loc = 'upper left'); # for legend to be displayed in our plot
```

```python
plot_forecast(y_train_IA,y_test_IA, arima_model)
```

### Check Stationarity

The time-series data must be stationary to develop a model that is effective in forecasting future values.  Firstly, let's plot Augmented Dickey–Fuller test (ADF test), the partial correlation function (PACF), and the autocorrelation function (ACF) to determine stationarity of the data.

```python
# Check stationarity:
ad_fuller_result = adfuller(y_train)
print(f"p-value for original: {str(ad_fuller_result[1])}")
```

##### Findings
From the ADF test, p-value is < 0.05. This tells us we can reject the null hypothesis (null hypothesis = the data is not stationary) because it is statistically insignificant. 

### Determining AR, MA values

ACF and PACF assume stationarity of the underlying time series. Now that we have confirmed our data is stationary, lets plot a ACF, PACF to determine our 'p' and 'q' values:

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))

# Plot ACF for different differencing orders
plot_acf(y_train.index, ax=ax1) 
ax1.set_title("ACF - Autocorrelation", fontsize=12)

plot_pacf(y_train.index, ax=ax2)
ax2.set_title("PACF - Partial Autocorrelation", fontsize=12)

plt.tight_layout()

# Show the plots
plt.show()
```

### Check Seasonality

```python
decomposition = seasonal_decompose(x=y_train, period=30, model='addicitve')

# Plot the decomposition
fig = decomposition.plot()
fig.set_size_inches((12,6))
fig.tight_layout();
```

```python
# Fit Auto ARIMA
arima_model = pm.AutoARIMA(
                           start_p=0, max_p=10, 
                           d=None,   
                           start_q=0, max_q=10, 
                           
                           start_P=2, max_P=3, 
                           D=0,    
                           start_Q=1, max_Q=10, 
                           m=30,     # daily sales value is used
                           seasonal=True, 
                           
                           # same parameters as Auto ARIMA
                           trace=True, 
                           random_state=20,
                           n_fits=50
                          )

arima_model.fit(y_train)
```

##### Inferences from Auto ARIMA model
- `AR` best hyperparameter:`p` - 1
- `I` best hyperparameter:`d` - 1
- `MA` best hyperparameter: `q` - 1

```python
# trained model summary stats
arima_model.summary()
```

```python
def evaluate_model(model, y_train, y_test):
    # Make predictions on train and test sets
    y_pred_train = model.predict(n_periods = len(y_train))
    y_pred_test = model.predict(n_periods = len(y_test))
    
    # Calculate RMSE and MAPE for train and test sets
    rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
    mape_train = mean_absolute_percentage_error(y_train, y_pred_train)
    mape_test = mean_absolute_percentage_error(y_test, y_pred_test)
    
    # Print the results
    print('Train RMSE:', rmse_train)
    print('Test RMSE:', rmse_test)
    print('Train MAPE:', mape_train)
    print('Test MAPE:', mape_test)
```
