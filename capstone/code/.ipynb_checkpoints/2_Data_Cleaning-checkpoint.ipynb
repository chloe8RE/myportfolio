## Data Inspection

### Function for dataset inspection

```python
def inspect_data(df):
    print(f'Data Dimension'.center(50,'='))
    print(f'The shape is: ', df.shape, '\n')  # print data dimension
    print(df.info())                  # check nulls and data types
    print(f'Duplicates'.center(50,'='))
    print(df.duplicated().value_counts())  # check duplicates
    print(f'Null values'.center(50,'='))
    null_counts = df.isnull().sum()    # check nulls
    print(null_counts[null_counts>0])  # Filter out the columns with nulls
    print(f'Null values %'.center(50,'='))
    print(null_counts[null_counts>0]/len(df))  # Calculate the percentage of columns with nulls
```

Perform an inspection on datasets on dimension, nulls, data types, duplicates.

```python
inspect_data(sales_df)
```

```python
sales_df.columns
```

Findings:
1. Inspection suggests no null values hence will keep all rows and not dropping.  But will check in detailed in next steps.
2. There is no duplicated rows. 
3. There is inconsistency in naming convention of the columns headers and will standardize it in next step.
4. `order_date`, `month` and `Sign in date` data types are *object*.  Will change to *datetime*.
5. `order_id` is *object*. Will check in detailed and change to *integer*.
6. `Zip` data type is *integer* in the dataset.  Setting it as *integer* is not ideal as it will eliminate zeros in front (e.g. <span style='color:blue'>**0**</span>2110) that are not meant to be kept for geographical identifier.  Hence will change to strings.

### Standardize columns

Standardize the columns naming convention by changing to lowercase and replace whitespaces with underscore _. 

```python
sales_df.columns = sales_df.columns.str.lower().str.replace(' ', '_')
```

```python
sales_df.head()
```

### Change date column to datetime

Change the columns `order_date`, `month` and `sign_in_date` with data types as *object* to *datetime*.  As these date columns are all in different formats, hence will change separately and print dataframe to confirm changes have been made.

```python
sales_df['order_date'] = pd.to_datetime(sales_df['order_date'],format='%d/%m/%Y')
```

```python
sales_df['month'] = pd.to_datetime(sales_df['month'],format='%b-%y').dt.strftime('%b-%Y')
```

```python
sales_df['sign_in_date'] = pd.to_datetime(sales_df['sign_in_date'],format='%m/%d/%Y')
```

```python
sales_df.head()
```

```python
sales_df[['order_date','month','sign_in_date']].info()
```

### Change string column to integer

Change the column `order_id` from *object* to *integer*.  Had this error message **'ValueError: invalid literal for int() with base 10: '100468520-1'** initially.  With this, decided to remove '-' in the `order_id` column and then successfully change the column from *object* to *integer*, just like the other id columns.

```python
sales_df['order_id'] = sales_df['order_id'].str.replace('-','',regex=True)
```

```python
sales_df['order_id'] = sales_df['order_id'].astype('int64')
```

```python
sales_df[['order_id','item_id','cust_id']].info()
```

### Rename columns

Change some of the columns header name to better represent the columns for ease of understanding.

```python
sales_df = sales_df.rename(columns={'status':'order_status',
                                    'qty_ordered':'order_qty',
                                    'total':'total_value',
                                    'phone_no._':'phone_num',
                                    'name_prefix':'title'})
```

```python
sales_df.head()
```

Check the mathematical accuracy of column `value` and `total_value`.  There are some discrepancy hence these features will not be used in modelling. Will follow up with the data provider to understand the reason of discrepancy

Noticed `discount_percent` do not match with `discount_amount` in 1574 rows. Further dive-down and noticed those are with empty value in column `discount_amount`.

```python
sales_df['discount_percent'] = sales_df['discount_amount'] / sales_df['value']
```

```python
sales_df['discount_is_correct'] = sales_df['discount_amount'] / sales_df['value'] == sales_df['discount_percent']
```
