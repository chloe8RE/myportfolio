### Check Stationarity

- Before ARIMA time series modelling, the first step to determine the number of differencing required to make the series stationary.
- Testing of stationarity always encompasses Augmented Dickey–Fuller test (ADF Test) and Kwiatkowski-Phillips-Schmidt-Shin test (KPSS test).
- First is to run ADF test. The null hypothesis of ACF test assumes time series is non-stationary. P-value derived should be LESS THAN significance level (say 0.05) to reject the null hypothesis.  Thereby inferring that the time series is stationary. 

```python
# Check stationarity using ADF test
ad_fuller = adfuller(x=mobiles_sales,regression='c')
print(f'p-value for ADF test: {str(ad_fuller[1])}')
```

- As above ADF test less than 0.05 significance level, hence null hypothesis of ADF test is rejected and I deduce mobiles & tablets data series is stationary.<br>

- Next is to run KPSS test to check the stationarity of series around a deterministic trend. A key difference from ADF test is the null hypothesis of the KPSS test is that the series is stationary, so the p-value interpretation is opposite to each other.  That said, if p-value is LESS THAN significance level (i.e. <0.05), then the time series data is non-stationary.

```python
# KPSS test is used to check stationarity with constant regression only.
kpss = kpss(x=mobiles_sales, regression='c')
print(f'p-value for KPSS test: {str(kpss[1])}')
```

From the above KPSS test, p-value is > 0.05, it suggests that there is evidence against the null hypothesis of stationarity around a deterministic trend.  **No differencing** is required for stationary time series.

```python
# ACF Plot
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))
plot_acf(mobiles_sales, ax=ax1) 
ax1.set_title("Autocorrelation - no differencing", fontsize=12)
plot_acf(mobiles_sales.diff().dropna(), ax=ax2)
ax2.set_title("Autocorrelation - 1st order differencing", fontsize=12)
plt.tight_layout()
plt.show()
```

Findings:
- To put concept of stationary into perspective, we can visualise the above ACF (Autocorrelation) plots. 
- First plot is without any differencing. Second plot with 1st order differencing and the 2nd lag goes into negative zone ~0.3 indicating over-differenced.  This is to confirm no differencing required which coincides with ADF and KPSS tests. In this case, d=0. 
- Now that I have checked the time series for stationary with constant mean and variance over time, it is suitable for modelling.

### Determine AR and MA

Next is to identify if model needs Auto-regressive (AR) terms (p) through Partial Autocorrelation Function (PACF) plot.

```python
# Plot ACF and PACF
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 6))
plot_acf(mobiles_sales, ax=ax1) 
ax1.set_title("Autocorrelation", fontsize=12)
plot_pacf(mobiles_sales, ax=ax2)
ax2.set_title("Partial Autocorrelation", fontsize=12)
plt.tight_layout()
plt.show()
```

Findings:
- Tailing off in ACF plot suggests geometric decay in AR modelling.
- From ACF plot (with no differencing) shows several spikes above the blue region reveals significant correlations which could be the MA 'q' value. Hence it can be further determined through AutoArima. 
- From PACF plot, there is high degree of autocorrelation at lag 1.  Lag 2 is within the significant threshold (blue region) hence p=1 for order of AR term.

### Model and Evaluation

Baseline Model - ARIMA

```python
# Fit Auto ARIMA to find the best p, d, q terms
arima_model = pm.AutoARIMA(
                           start_p=0, max_p=10, 
                           d=0,   
                           start_q=0, max_q=10, 

                           trace=True, 
                           random_state=20,
                           n_fits=50
                          )

arima_model.fit(y_train)
```

##### Inferences from Auto ARIMA model
- `AR` best hyperparameter:`p` - 1
- `I` best hyperparameter:`d` - 0
- `MA` best hyperparameter: `q` - 1

```python
# trained model summary stats
arima_model.summary()
```

#### Function for model evaluation

```python
def evaluate_model(model, y_train, y_test):
    # Make predictions on train and test sets
    y_pred_train = model.predict(n_periods = len(y_train))
    y_pred_in_sample = model.predict_in_sample()
    y_pred_test = model.predict(n_periods = len(y_test))
    
    # Calculate RMSE and MAPE for train and test sets
    rmse_train = np.sqrt(mean_squared_error(y_train, y_pred_train))
    rmse_train_in_sample = np.sqrt(mean_squared_error(y_train, y_pred_in_sample))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))
    
    # Print the results
    print('Train RMSE:', rmse_train)
    print('Train RMSE in sample:', rmse_train_in_sample)
    print('Test RMSE:', rmse_test)
```

```python
evaluate_model(arima_model, y_train, y_test)
```

#### Function for plotting

```python
def plot_forecast(y_train, y_test, model, X_test=None):
    # Generate len(y_test) number of predictions
    y_pred = model.predict(n_periods = len(y_test), X=X_test)

    plt.figure(figsize=(20,8))

    # Plot training data
    plt.plot(y_train, color='blue', label='y_train')

    # Plot testing data. 
    plt.plot(y_test, color='orange', label='y_test', alpha = 0.7)

    # Plot predicted test values
    plt.plot(y_pred, color='green', label='y_pred', alpha = 0.9, ls = '--')

    y_pred_in_sample = model.predict_in_sample()
    y_pred_test = model.predict(n_periods = len(y_test))
    
    # Find the MAPE (mean abs percentage error) of the predictions
    rmse_train= np.sqrt(mean_squared_error(y_train, y_pred_in_sample))
    rmse_test = np.sqrt(mean_squared_error(y_test, y_pred_test))

    plt.title(label = f'Mobiles & Tablets - Sales Quantity Predictions\n\n Train RMSE:{rmse_train:.2f};  Test RMSE: {rmse_test:.2f}', fontsize=12)
    plt.legend(fontsize = 12, loc = 'upper left'); 
```

```python
plot_forecast(y_train,y_test, arima_model)
```

Findings from ARIMA model: <br>
1. Per observation in the above plot, my baseline model, ARIMA overfits and shows a flatline prediction. This happens could be possibly the historical data doesn't have strong seasonality and the ARIMA model finds difficult to predict the future data points therefore it simply take average of previous values and predict as future.
2. Another reason could be time series data has too much noises and irregularities which is hard for ARIMA model to find a stable relationship hence it simply predicted faltline.

## Time Series - ARIMAX Model

Baseline model is established based on ARIMA not showing an ideal result, I proceed to use multivariate time series forecasting model, ARIMAX, to try model for future order quantity with exogenous features to see if any improvement. 
1. ARIMAX is an extension of traditional ARIMA to include both past values of the responses and additional features to predict future values of the response to improve the model's predictive power.
2. The 'X' in ARIMAX accounts for 'exogenous' i.e. external features.
3. 'S' seasonal component will be left out to isolate performance without any seasonality component

### Preprocessing data

Filter for completed sales by category = 'Mobiles & Tablets'

```python
mobiles_sales_exog = filtered_sales[filtered_sales['category']=='Mobiles & Tablets']
```

Select external features that is exogenous 

```python
mobiles_sales_exog = mobiles_sales_exog[['order_qty','price','discount_amount',
                                       'payment_method','full_name','gender',
                                       'age','state','region']]
```

Create function for most frequent values for categorical columns

```python
def most_frequent_mode(series):
    mode_result = series.mode()
    return mode_result[0] if not mode_result.empty else None
```

Resample the dataset to daily and aggregate function according to the data types of each column.  Then fill NaN values by forward-fill the last valid observation to next values.

```python
mobiles_sales_exog = mobiles_sales_exog.resample('D').agg({'order_qty': 'sum',
                                                         'price': 'mean',
                                                         'discount_amount': 'mean',
                                                         'payment_method': most_frequent_mode,
                                                         'full_name': most_frequent_mode,
                                                         'gender': most_frequent_mode,
                                                         'age': 'mean',
                                                         'state': most_frequent_mode,
                                                         'region': most_frequent_mode}).fillna(method='ffill')
```

```python
# Check results
mobiles_sales_exog
```

Per observation above, the dataset ends 2021-08-29 hence reindex to 2021-09-30 to get full 365 days for modelling.  Then fill NaN values by forward-fill the last valid observation to next values

```python
mobiles_sales_exog = mobiles_sales_exog.reindex(new_index, method='ffill')
```

```python
# Check results
mobiles_sales_exog
```

### Setting up X features

```python
mobiles_sales_exo = mobiles_sales_exog[mobiles_sales_exog.index<='2021-06-30']
mobiles_sales_exo
```

Split X features into numericals and categoricals

```python
X_num = mobiles_sales_exo[['price','discount_amount','age']]
X_cat = mobiles_sales_exo[['payment_method','gender','state','region']]
```

Train Test Split

```python
X_num_train = X_num[:182]
X_num_test = X_num[182:]
X_cat_train = X_cat[:182]
X_cat_test = X_cat[182:]
```

```python
ohe = OneHotEncoder(handle_unknown='infrequent_if_exist', sparse_output=False)
X_cat_train_encoded = pd.DataFrame(ohe.fit_transform(X_cat_train), columns=ohe.get_feature_names_out())
X_cat_test_encoded = pd.DataFrame(ohe.transform(X_cat_test), columns=ohe.get_feature_names_out())
```

```python
X_cat_train_encoded
```

```python
X_cat_test_encoded
```

### StandardScaler for numerical features

```python
ss = StandardScaler()
X_num_train_scaled = ss.fit_transform(X_num_train)
X_num_test_scaled = ss.transform(X_num_test)
X_num_train_scaled = pd.DataFrame(X_num_train_scaled, columns=X_num_train.columns)
X_num_test_scaled = pd.DataFrame(X_num_test_scaled, columns=X_num_test.columns)
```

```python
X_num_train_scaled
```

```python
X_num_test_scaled
```

Merge the categoricals and numericals 

```python
Z_train = pd.merge(X_num_train_scaled, X_cat_train_encoded, left_index=True, right_index=True)
Z_test = pd.merge(X_num_test_scaled, X_cat_test_encoded, left_index=True, right_index=True)
```

```python
Z_train
```

Assign the datetime index back to the merged dataframes

```python
mobiles_sales_exo.index
```

```python
Z_train.index = mobiles_sales_exo.index[:182]
Z_test.index = mobiles_sales_exo.index[182:]
```

```python
Z_train
```

### Model and Evaluation

Fit ARIMAX Model with Exogenous Features

```python
arimax_model = pm.auto_arima(y_train, X=Z_train, 
                            start_p=1, max_p=50,
                            d=None, D=None,
                            start_q=1, max_q=50,
                            seasonal=False,
                            trace=True,
                            random_state=1,
                            n_fits=100)
```

#### Function to plot forecast with the input for period of forecast

```python
def plot_forecast_period(y_train, y_test, arima_model, period, X_test=None):
    # Generate len(y_test) number of predictions
    y_pred = arima_model.predict(n_periods = period, X=X_test[:period])

    plt.figure(figsize=(20,6))

    # Plot training data
    plt.plot(y_train, color='blue', label='y_train')

    # Plot testing data. 
    plt.plot(y_test, color='orange', label='y_test', alpha = 0.7)

    # Plot predicted test values
    plt.plot(y_pred, color='green', label='y_pred', alpha = 0.9, ls = '--')

    y_pred_in_sample = arima_model.predict_in_sample(X_test)
    y_pred_test = arima_model.predict(n_periods = period,X=X_test[:period])
    
    # Find the MAPE (mean abs percentage error) of the predictions
    rmse_train= np.sqrt(mean_squared_error(y_train, y_pred_in_sample))
    rmse_test = np.sqrt(mean_squared_error(y_test[:period], y_pred_test))

    plt.title(label = f'Mobiles & Tablets - Sales Quantity Predictions\n\n Train RMSE:{rmse_train:.2f};  Test RMSE: {rmse_test:.2f}', fontsize=12)
    plt.legend(fontsize = 12, loc = 'upper left'); 
```

```python
plot_forecast_period(y_train, y_test, arimax_model, 91, Z_test)
```

```python
plot_forecast_period(y_train, y_test, arimax_model, 60, Z_test)
```

## Time Series - SARIMAX Model

Fit SARIMAX Model with Exogenous Features

```python
sarimax_model = pm.auto_arima(y_train, X=Z_train, 
                            start_p=1, max_p=50,
                            d=None, 
                            start_q=1, max_q=50,
                            start_P=1, max_P=50,
                            D=None,
                            start_Q=1, max_Q=50,
                            m=7,
                            seasonal=True,
                            trace=True,
                            random_state=1,
                            n_fits=100)
```

```python
plot_forecast_period(y_train, y_test, sarimax_model, 91, Z_test)
```

Findings: 
1. Based on model performance evaluation, SARIMAX model is less overfitting than SARIMA model with lesser differential values between train and test RMSE as compared to ARIMA/ARIMAX models. 
2. SARIMAX model yields better results, demonstrates stronger predictive capability for short term 60 to 90 days of foreacasting window.
3. Per observation on linegraph above, SARIMAX model is able to predict the uptrend within these short intervals. The spikes in demand lies in the quarter Oct-2020 to Dec-2020 with max order quantity of 4,261 units hence RMSE 196 is 4.6% relative forecasting errors to max order quantity.
4. Short term forecasting face more challenges than longer term forecasting because it is sensitive to recent fluctuations and unforeseen events. Therefore the RMSE derived from SARIMAX model is considered acceptable for short-term forecasting with expected variablity.
6. Short term foreacasting is also useful for highly competitive and fast-paced ecommerce industry, to be adopted in conjuction with long term forecasting as immediate changes can be made to the models to get quick and relatively accutate results. Changing long term forecasting may involve longer processes hence it is not advisable because meeting customer demand is upmost importance for business longevity.
